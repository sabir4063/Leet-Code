## Pairwise swap elements of a linked list by swapping data

```python
class Node(object):
    
    def __init__(self, x):
        
        self.val = x
        self.next = None
       
    
class LL(object):
    
    def __init__(self):
        
        self.head = None
        
        
    def pairWiseSwap(self, head):
        # code here

        prev = None
        f = head
        s = head.next
        new_head = None

        while s:
            if s.next:
                temp = s.next
            else:
                temp = None
            if s.next and s.next.next:
                temp2 = s.next.next
            else:
                temp2 = None
            if prev is None:
                prev = f
                new_head = s
                s.next = f
                f.next = temp
            else:
                prev.next = s
                s.next = f
                f.next = temp
                prev = f
            f = temp
            s = temp2
        head = new_head
        return head
```
