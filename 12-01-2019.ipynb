```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        
        count = 0 
        
        if head is None: return
        
        curr = head
        while curr:
            count += 1
            curr = curr.next
        
        if n <= 0 or n > count:
            return head
        
        ind = count - n 
        count = 0
        prev = None
        curr = head
        
        while curr:
            if ind == count:
                break
            count += 1
            prev = curr
            curr = curr.next
        
        if prev == None:
            head = head.next
        else:
            prev.next = curr.next
        
        return head
```

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def reorderList(self, head):
        """
        :type head: ListNode
        :rtype: void Do not return anything, modify head in-place instead.
        """
        
        if not head or not head.next or not head.next.next: return
        
        slow = head
        fast = head
        
        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next

        mid_head = slow.next
        prev = None
        curr = mid_head
        slow.next = None
        
        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
            
        mid_head = prev
        curr1 = head
        curr2 = mid_head
        
        while curr2:
            t1 = curr1.next
            t2 = curr2.next
            curr1.next = curr2
            curr2.next = t1
            curr1 = t1
            curr2 = t2
               
```
