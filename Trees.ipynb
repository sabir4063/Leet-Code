```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def isUnivalTree(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        
        vals = []
        def inorder(root):
            if root:
                vals.append(root.val)
                
                temp = root.left
                inorder(temp)

                temp = root.right
                inorder(temp)

            return len(list(set(vals)))
        
        v = inorder(root)
        #print(vals)
        if v==1:
            return True
        return False
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        
        def dfs(root):
            if root is None:
                return 0
    
            l = dfs(root.left)
            r = dfs(root.right)
           
            return max(l,r)+1
        
        return dfs(root)
```

```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, children):
        self.val = val
        self.children = children
"""
class Solution(object):
    def preorder(self, root):
        """
        :type root: Node
        :rtype: List[int]
        """
        
        if root is None:
            return []
        
        ans = []
        stk = []
        stk.append(root)
        
        while stk:
            temp = stk.pop()
            ans.append(temp.val)
            stk.extend(temp.children[::-1])
            
        return ans
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def invertTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        
        if root==None: return None
        
        left = self.invertTree(root.left)
        right = self.invertTree(root.right)
        root.left = right
        root.right = left
        
        return root
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def diameterOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.dia = 0
        
        def traverse(root):
            if root is None:
                return 0
            else:
                l = traverse(root.left)
                r = traverse(root.right)
                self.dia = max(self.dia, l+r)
                #print(root.val, l, r)
                return max(l,r)+1
            
        traverse(root)
        return self.dia
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
    
        def left_traverse(root, vals):
            if root:
                vals.append(root.val)
                left_traverse(root.left, vals)
                left_traverse(root.right, vals)
            else:
                vals.append(None)
            return vals
        
        
        def right_traverse(root, vals):
            if root:
                vals.append(root.val)
                right_traverse(root.right, vals)
                right_traverse(root.left, vals)
            else:
                vals.append(None)
            return vals        
                
        l = left_traverse(root, [])
        r = right_traverse(root, [])
        
        #print(l,r)
        
        if l==r: return True
        else: return False
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isSubtree(self, s, t):
        """
        :type s: TreeNode
        :type t: TreeNode
        :rtype: bool
        """
        
        def pre_traverse(root, vals):
            if root:
                vals.append(root.val)
                pre_traverse(root.left, vals)
                pre_traverse(root.right, vals)
            else:
                vals.append(None)
            return vals
        
            
        def search_sub_tree(root):
            if root:
                temp = pre_traverse(root, [])
                if self.t == temp: 
                    self.tag = True
                search_sub_tree(root.left)
                search_sub_tree(root.right)
            
                
        self.tag = False        
        self.t = pre_traverse(t, [])
        search_sub_tree(s)
        
        return self.tag
```
