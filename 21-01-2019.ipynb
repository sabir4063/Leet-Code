```python
class Solution:
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """
        
        def permutation(tmp,l,r):
            #print(tmp)
            if len(tmp)==2*n:
                res.append(tmp)
                return
            
            if l < n:
                permutation(tmp+'(', l+1, r)
            if r < l:
                permutation(tmp+')', l, r+1)
                    
        res = []
        permutation('',0,0)
        
        return res
```

```python
class Solution:
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """
        
        nums = [i for i in range(1,n+1)]
        #print(nums)
        res = []
        visited = [False]*(n+1)
        
        def backtrack(temp,s,e):
            
            if len(temp)==k:
                res.append(temp[:])
                return
            
            for i in range(s,e):
                if visited[i]==True: continue
                    
                visited[i]=True    
                temp.append(nums[i])
                backtrack(temp,i+1,e)
                temp.pop()
                visited[i]=False
                
        backtrack([],0,n)
        
        return res
                
```

```python
class Solution:
    def permuteUnique(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        
        nums.sort()
        res=[]
        size = len(nums)
        visited = [False]*size
        
        def backtrack(temp, s, e):
            
            if len(temp)==size:
                res.append(temp[:])
                
            for i in range(0,e):
                if visited[i]: 
                    continue
                if (i>0 and nums[i]==nums[i-1] and visited[i-1]==False): 
                    continue
                visited[i]=True
                temp.append(nums[i])
                backtrack(temp,i+1,e)
                visited[i]=False
                temp.pop()

        backtrack([],0,size)
        
        return res
```
