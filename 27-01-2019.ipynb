```python
class Solution:
    def pacificAtlantic(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[List[int]]
        """
        
        if matrix==[]: return []
        
        res = []
        r = len(matrix)
        c = len(matrix[0])
        
        res = []
        
        def is_valid(i,j):
            if i>=0 and i<r and j>=0 and j<c:
                return True
            return False
                
            
        def dfs_pacific(i,j,visited):
            
            if self.p_visited[i][j]==False:
                self.p_visited[i][j]=True
            
                for adj in [(i+1,j),(i,j+1),(i-1,j),(i,j-1)]:
                    x,y=adj
                    if(is_valid(x,y)) and (x,y) not in visited and matrix[x][y]>=matrix[i][j]:
                        dfs_pacific(x,y,visited)
                        

        def dfs_atlantic(i,j,visited):
            
            if self.a_visited[i][j]==False:
                self.a_visited[i][j]=True
            
                for adj in [(i+1,j),(i,j+1),(i-1,j),(i,j-1)]:
                    x,y=adj
                    if(is_valid(x,y)) and (x,y) not in visited and matrix[x][y]>=matrix[i][j]:
                        dfs_atlantic(x,y,visited)
                        
        
        self.p_visited = [[False] * c for _ in range(r)]
        self.a_visited = [[False] * c for _ in range(r)]
        
        for i in range(c):
            dfs_pacific(0,i,[])
            dfs_atlantic(r-1,i,[])
                
        for i in range(r):
            dfs_pacific(i,0,[])    
            dfs_atlantic(i,c-1,[])
    
        for i in range(r):
            for j in range(c):
                if self.p_visited[i][j] and self.a_visited[i][j]:
                    res.append([i,j])
                    
        return res
```

```python
class Solution:
    def updateMatrix(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[List[int]]
        """
        
        q=collections.deque()
        r=len(matrix)
        c=len(matrix[0])
        visited=[[False] * c for _ in range(r)]
        
        for i in range(r):
            for j in range(c):
                if matrix[i][j]==0:
                    visited[i][j]=True
                    q.append((i,j))
                    
        while q:
            x,y=q.popleft()
            for i,j in [(x+1,y),(x-1,y),(x,y+1),(x,y-1)]:
                if i>=0 and j>=0 and i<r and j<c:
                    if visited[i][j]==False:
                        visited[i][j]=True
                        matrix[i][j]=matrix[x][y]+1
                        q.append((i,j))
                        
                        
        return matrix
        
        
```

```python
class Solution:
    def findOrder(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: List[int]
        """
        
        graph=collections.defaultdict(list)
        
        for edge in prerequisites:
            graph[edge[0]].append(edge[1])
            
        res = []
        self.ans=False
        visited = [0]*numCourses
        
        
        def topological_sort(node):
            visited[node]=1
            for v in graph[node]:
                if visited[v]==1:
                    self.ans=True
                    return
                elif visited[v]==0:
                    topological_sort(v)
                    
            visited[node]=2
            res.append(node)
            
        
        for i in range(numCourses):
            if visited[i]==0:
                topological_sort(i)
            if self.ans==True:
                return []
                
        return res
```

```python
class Solution:
    def findOrder(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: List[int]
        """
        
        graph=collections.defaultdict(list)
        
        for edge in prerequisites:
            graph[edge[0]].append(edge[1])
            
        res = []
        in_degree=[0]*numCourses
        q = []
        
        for i in range(numCourses):
            for e in graph[i]:
                in_degree[e]+=1
                
        for i in range(numCourses):
            if in_degree[i]==0:
                q.append(i)
        
        while q:
            node = q.pop(0)
            for e in graph[node]:
                in_degree[e]-=1
                if in_degree[e]==0:
                    q.append(e)
            
            res.append(node)
            
        if len(res)==numCourses:
            return res[::-1]
        else:
            return []
```
