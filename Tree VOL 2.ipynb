```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if root == None: return []
        
        q = collections.deque()
        depth = 1
        q.append([root,depth])
        res = {}
        res[depth] = [root.val]
        
        while q:
            node, depth = q.popleft()
            depth += 1
            temp = []
            if node.left:
                q.append([node.left, depth])
                temp.append(node.left.val)
            if node.right:
                q.append([node.right, depth])
                temp.append(node.right.val)
            
            if temp != []:
                if depth in res:
                    t = res[depth]
                    t.extend(temp)
                    res[depth] = t
                else:
                    res[depth] = temp
                    
        ans = []
        for i in range(1, depth):
            ans.append(res[i])
        
        return ans
            
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def trimBST(self, root, L, R):
        """
        :type root: TreeNode
        :type L: int
        :type R: int
        :rtype: TreeNode
        """
        
        if root is None: return None
        
        def traverse(root):
            
            if not root: 
                return None
            elif root.val<L:
                return traverse(root.right)
            elif root.val>R:
                return traverse(root.left)
            else:
                root.left = traverse(root.left)
                root.right = traverse(root.right)
                return root
            
                
        return traverse(root)
                    
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isBalanced(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        
        if root == None: return True
        
        def traverse(root, depth):
        
            if root is None: return depth
            
            l = traverse(root.left, depth+1)
            r = traverse(root.right, depth+1)
                
            if l==-1 or r==-1 or abs(l-r)>1: return -1
            
            return max(l,r)
    
        
        if traverse(root, 0)!=-1: 
            return True 
        else: 
            return False        
        
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def findTarget(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: bool
        """
        
        q = collections.deque()
        q.append(root)
        s = set()
        
        while q:
            node = q.popleft()
            rem = k-node.val
            if rem in s:
                return True
            s.add(node.val)
            
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
                
        
        return False
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def pathSum(self, root, sum):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: List[List[int]]
        """
        
        if root is None: return []
        
        path = {}
        path[root] = None
        res = []
        
        def dfs(root, cost):
            
            if cost==sum and root.left==None and root.right==None:
                temp = root
                s = []
                while(temp):
                    s.append(temp.val)
                    temp = path[temp]
                res.append(s[::-1])
                return
            
            if root.left:
                path[root.left] = root
                dfs(root.left, cost+root.left.val)
            if root.right:
                path[root.right] = root
                dfs(root.right, cost+root.right.val)
                
        dfs(root, root.val)
        
        return res
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def findSecondMinimumValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        
        if root is None: return -1
        
        s = set()
        
        def dfs(root):
            if not root: return
            s.add(root.val)
            dfs(root.left)
            dfs(root.right)
            
        
        dfs(root)
        if len(s)==1: return -1
        
        mn1 = sorted(list(s))[1]
        return mn1
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if root is None: return None
        
        self.ans = None
        
        def dfs(root,p,q):
            
            if root is None: return 
            
            
            if p.val < root.val and q.val < root.val:
                dfs(root.left,p,q)
            elif p.val > root.val and q.val > root.val:
                dfs(root.right,p,q)
            else:
                self.ans = root
                return 
        
        dfs(root,p,q)
        return self.ans
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None: return None
    
        q = collections.deque()
        depth = 1
        res = {}
        q.append([root,depth])
        
        while q:
            node, depth = q.popleft()
            if depth not in res:
                res[depth] = node.val
            
            depth+=1
            if node.left:
                q.append([node.left, depth])
            if node.right:
                q.append([node.right, depth])
        
        return res[depth-1]
        
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def zigzagLevelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        
        if root==None: return []
        
        q = collections.deque()
        depth = 1
        q.append([root,depth])
        path = {}
        
        while q:
            node, depth = q.popleft()
            if depth in path:
                temp = path[depth]
            else:
                temp = []
            temp.append(node.val)
            path[depth] = temp
            depth+=1
            
            if node.left:
                q.append([node.left, depth])
            if node.right:
                q.append([node.right, depth])
                
        res = []
        for k in range(1,depth):
            if k%2==0:
                res.append(path[k][::-1])
            else:
                res.append(path[k])
        
        return res
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def flipEquiv(self, root1, root2):
        """
        :type root1: TreeNode
        :type root2: TreeNode
        :rtype: bool
        """
        if root1==None and root2==None: return True
        if root1==None or root2==None: return False
        
        self.tag = True
        
        def search(root1, root2):
            if root1 and root2:
                #print(root1.val, root2.val, root1.right, root1.left, root2.right, root2.left, self.tag)
                if root1.val==root2.val:
                    if root1.left and root2.left and root1.right and root2.right:
                        if root1.left.val == root2.left.val and root1.right.val == root2.right.val:
                            search(root1.left, root2.left)
                            search(root1.right, root2.right)

                        elif root1.left.val == root2.right.val and root1.right.val == root2.left.val:
                            temp = root2.left
                            root2.left = root2.right
                            root2.right = temp    
                            search(root1.left, root2.left)
                            search(root1.right, root2.right)
                        else:
                            self.tag = False
                            return

                    elif root1.left and root2.left and (root1.right is None and root2.right is None):
                        if root1.left.val == root2.left.val:
                            search(root1.left, root2.left)
                        else:
                            self.tag = False
                            return

                    elif root1.right and root2.right and (root1.left is None and root2.left is None):
                        if root1.right.val == root2.right.val:
                            search(root1.right, root2.right)
                        else:
                            self.tag = False
                            return
                        
                    elif root1.left and root2.right and (root1.right is None and root2.left is None):
                        if root1.left.val == root2.right.val:
                            root2.left = root2.right
                            root2.right = None    
                            search(root1.left, root2.left)
                        else:
                            self.tag = False
                            return
                    
                    elif root1.right and root2.left and (root1.left is None and root2.right is None):
                        if root1.right.val == root2.left.val:
                            root2.right = root2.left
                            root2.left = None    
                            search(root1.right, root2.right)
                        else:
                            self.tag = False
                            return
                        
                    elif (root1.left and root1.right) and (root2.left==None or root2.right==None):
                        #print('a')
                        self.tag = False
                        return
                    
                    elif (root2.left and root2.right) and (root1.left==None or root1.right==None):
                        #print('b')
                        self.tag = False
                        return
                    
                    elif (root1.left or root1.right) and (root2.left==None and root2.right==None):
                        #print('c')
                        self.tag = False
                        return
                    
                    elif (root2.left or root2.right) and (root1.left==None and root1.right==None):
                        #print('d')
                        self.tag = False
                        return
                    
                    else:
                        return
                else:
                    self.tag = False
                    return
                
        search(root1, root2)
        return self.tag
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root == None: return []
        
        q = collections.deque()
        depth = 1
        path = {}
        q.append([root,depth])
        
        while q:
            node, depth = q.popleft()
            if depth in path:
                t = path[depth]
            else:
                t = []
            t.append(node.val)
            path[depth] = t
            depth+=1
            
            if node.left:
                q.append([node.left, depth])
                
            if node.right:
                q.append([node.right, depth])
                
        ans = []
        for i in range(1, depth):
            ans.append(max(path[i]))
            
        return ans
```
