```python
# Definition for binary tree with next pointer.
# class TreeLinkNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
#         self.next = None

class Solution:
    # @param root, a tree link node
    # @return nothing
    def connect(self, root):

        if root == None: return 

        q = collections.deque()
        depth = 1
        path = {}
        q.append([root,depth])

        while q:
            node, depth = q.popleft()

            if depth in path:
                t = path[depth]
                t.next = node

            path[depth] = node
            depth+=1

            if node.left:
                q.append([node.left, depth])

            if node.right:
                q.append([node.right, depth])
        
        
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def sortedArrayToBST(self, nums):
        """
        :type nums: List[int]
        :rtype: TreeNode
        """
        
        def to_bst(nums):
            
            if nums==[]:
                return None
            
            m = len(nums)//2
            root = TreeNode(nums[m])
            root.left = to_bst(nums[:m])
            root.right = to_bst(nums[m+1:])
            
            return root
    
        return to_bst(nums)
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def widthOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        
        if not root: return 0
    
        depth = collections.defaultdict(list)
        d = 1
        q = collections.deque()
        cnt = 1
        q.append([root, cnt, d])
        
        while q:
            node, cnt, d = q.popleft()
            depth[d].append(cnt)
            
            d+=1
            if node.left:
                q.append([node.left, cnt*2, d])
            
            if node.right:
                q.append([node.right, cnt*2+1, d])
        
        mx = 0
        for i in range(1,d):
            temp = max(depth[i]) - min(depth[i]) + 1
            if temp>mx:
                mx = temp
        return mx
            
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        
        def find_path( root, path, target):
            
            if root is None: return
            if self.tag==True: return
            path.append(root)
            
            if root.val == target:
                self.tag = True
                return
            
            if root.left:
                find_path(root.left, path, target)
            if root.right:
                find_path(root.right, path, target)
            
            if self.tag==True: return
            
            path.pop()
            return
        
        
        path1 = []
        path2 = []
        self.tag = False
        find_path(root,path1,p.val)
        self.tag = False
        find_path(root,path2,q.val)
        
        i = 0 
        while(i < len(path1) and i < len(path2)): 
            if path1[i].val != path2[i].val: 
                break
            i += 1
        return path1[i-1] 
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def distanceK(self, root, target, K):
        """
        :type root: TreeNode
        :type target: TreeNode
        :type K: int
        :rtype: List[int]
        """
        path = collections.defaultdict(list)
        self.t = None
        
        def dfs(root, target):
            
            if not root: return
            if root==target:
                self.t=target
                
            if root.left:
                path[root].append(root.left)
                path[root.left].append(root)
                dfs(root.left, target)
                
            if root.right:
                path[root].append(root.right)
                path[root.right].append(root)
                dfs(root.right, target)
            
            return
        
        dfs(root, target)
        d = 0
        q = collections.deque()
        q.append([target, d])
        res = collections.defaultdict(list)
        res[0].append(target.val)
        v = {}
        
        while q:
            node, d = q.popleft()
            v[node]=True
            d+=1
            
            for c in path[node]:
                if c not in v:
                    #print(d, c.val)
                    q.append([c,d])
                    res[d].append(c.val)
        #print(res)   
        return res[K]
```

```python
class Solution(object):
    def numTrees(self, n):
        """
        :type n: int
        :rtype: int
        """
        
        def catalan(n): 
            # Base Case 
            if n <=1 : 
                return 1 

            # Catalan(n) is the sum of catalan(i)*catalan(n-i-1) 
            C = 1
            for i in range(0, n):
                C = C * 2*(2*i+1)/(i+2)
            return int(C)
        
        return catalan(n)
```
