```python
class Solution:
    def grayCode(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        
        if n==0: return [0]
        
        res = []
        res.append('0')
        res.append('1')
        
        temp = res[:]
        ans = []
        
        for i in range(1,n):
            res = temp[:]
            temp=[]
            for j in range(len(res)):
                t = '0'+res[j]
                temp.append(t)
                
            for j in range(len(res)-1,-1,-1):
                t = '1'+res[j]
                temp.append(t)
        
        for t in temp:
            ans.append((int(t,2)))
        
        return ans
```

```python
class Solution:
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        
        res = []
        
        def backtrack(s, temp):
            if not s:
                res.append(temp[:])
                return
            
            for i in range(1,len(s)+1):
                if s[:i]==s[:i][::-1]:
                    backtrack(s[i:],temp+[s[:i]])
                    
        backtrack(s,[])
            
        return res
```

```python
class Solution:
    def floodFill(self, image, sr, sc, newColor):
        """
        :type image: List[List[int]]
        :type sr: int
        :type sc: int
        :type newColor: int
        :rtype: List[List[int]]
        """
        
        def dfs(i,j,r,c,seed,newColor):
            image[i][j]=newColor
            for xy in [(i+1,j),(i-1,j),(i,j+1),(i,j-1)]:
                x = xy[0]
                y = xy[1]
                
                if x>=0 and x<r and y>=0 and y<c and image[x][y]==seed and visited[x][y]==False:
                    visited[x][y]=True
                    dfs(x,y,r,c,seed,newColor)
                    visited[x][y]=False
                    
                    
        r = len(image)
        c = len(image[0])
        seed=image[sr][sc]
        visited = []
        for i in range(r):
            t = [False]*c
            visited.append(t)
        dfs(sr,sc,r,c,seed,newColor)
        
        return image
                    
```

```python
class Solution:
    def networkDelayTime(self, times, N, K):
        """
        :type times: List[List[int]]
        :type N: int
        :type K: int
        :rtype: int
        """
        
        dist = {}
        
        for i in range(1,N+1):
            for j in range(1,N+1):
                dist[(i,j)]=9999
        
        for time in times:
            i,j,t=time[0],time[1],time[2]
            dist[(i,j)]=t
        
        for k in range(1,N+1): 
            for i in range(1,N+1): 
                for j in range(1,N+1): 
                    dist[(i,j)] = min(dist[(i,j)],dist[(i,k)]+ dist[(k,j)])
        
        s = -1
        
        for i in range(1,N+1):
            if K!=i and dist[(K,i)]==9999: return -1
            if K!=i and dist[(K,i)]<9999:
                s=max(s,dist[(K,i)])
                    
        return s
        
        
```

```python
class Solution:
    def canVisitAllRooms(self, rooms):
        """
        :type rooms: List[List[int]]
        :rtype: bool
        """
        
        seen=[0]*len(rooms)
        
        def dfs(s,cnt):
            seen[s]=1
            
            for adj in rooms[s]:
                if visited[adj]==False:
                    visited[adj]=True
                    dfs(adj,cnt)
                    
        
        visited=[False]*len(rooms)
        visited[0]=True
        dfs(0,1)
        
        return len(rooms)==sum(seen)
```

```python
class Solution:
    def findCircleNum(self, M):
        """
        :type M: List[List[int]]
        :rtype: int
        """
        
        size = len(M)
        seen = [False]*size
        
        
        def dfs(node):
            seen[node]=True
                    
            for i in range(size):
                if seen[i]==False and M[node][i]==1:
                    dfs(i)
            
            
        cnt = 0
        for i in range(size):
            if seen[i]==False:
                cnt+=1
                dfs(i)
                
        return cnt
```

```python
class Solution:
    def shortestBridge(self, A):
        """
        :type A: List[List[int]]
        :rtype: int
        """
                
        depth = 0
        q=collections.deque()

        def alter_number(i,j):
            for x,y, in [(i+1,j),(i-1,j),(i,j+1),(i,j-1)]:
                if x>=0 and x<r and y>=0 and y<c:
                    if A[x][y]==1:
                        A[x][y]=2
                        q.append((x,y,0))
                        alter_number(x,y)
     
    
        r=len(A)
        c=len(A[0])
        tag=False
        for i in range(r):
            for j in range(c):
                if A[i][j]==1:
                    tag=True
                    A[i][j]=2
                    q.append((i,j,0))
                    alter_number(i,j)
                    break
            if tag: break
        
        
            
            
        #print(A)
        self.mn = 110
        while q:
            i,j,depth=q.popleft()
            #print(A)
            for x,y, in [(i+1,j),(i-1,j),(i,j+1),(i,j-1)]:
                if x>=0 and x<r and y>=0 and y<c:
                    if A[x][y]==0:
                        A[x][y]=3
                        q.append((x,y,depth+1))
                    if A[x][y]==1:
                        #print(depth,x,y)
                        if self.mn>depth:
                            self.mn=depth
        
        
        return self.mn
                        
```

```python
class Solution:
    def isBipartite(self, graph):
        """
        :type graph: List[List[int]]
        :rtype: bool
        """
        
        par={}
        col={}
        vst={}
        q=collections.deque()
        
        for root in range(len(graph)):
            if root not in vst:
                q.append(root)
                par[root]=root
                col[root]=1
                vst[root]=True

                while q:
                    node=q.popleft()
                    for e in graph[node]:
                        if e not in vst:
                            vst[e]=True
                            par[e]=node
                            col[e]=0 if col[node]==1 else 1
                            q.append(e)
                        else:
                            if col[e]==col[node]: 
                                return False
        
        return True
            
```

```python
class Solution:
    def possibleBipartition(self, N, dislikes):
        """
        :type N: int
        :type dislikes: List[List[int]]
        :rtype: bool
        """
        
        graph = collections.defaultdict(list)
        for u, v in dislikes:
            graph[u].append(v)
            graph[v].append(u)
        
        vst={}
        par={}
        clr={}
        q=collections.deque()
                
        for i in range(1,N+1):
            if i not in vst:
                q.append(i)
                vst[i]=1
                par[i]=1
                clr[i]=1
                while q:
                    node = q.popleft()
                    for edge in graph[node]:
                        e=edge
                        if e not in vst:
                            vst[e]=1
                            par[e]=node
                            clr[e]=0 if clr[node]==1 else 1
                            q.append(e)
                        else:
                            if clr[e]==clr[node]:
                                return False
                        
        return True
                    
```
