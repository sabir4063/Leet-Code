```python
class Solution:
    def findWords(self, board, words):
        """
        :type board: List[List[str]]
        :type words: List[str]
        :rtype: List[str]
        """
        
        r = len(board)
        c = len(board[0])
        
        trie={}
        for w in words:
            t=trie
            for ch in w:
                if ch not in t:
                    t[ch]={}
                t=t[ch]
            t['#']='#'
            
        #print(trie)
        
        
        def dfs(x,y,t,res):
            #print(board[x][y], t, res)
            
            if '#' in t:
                if res not in ans:
                    ans.append(res[:])
            
            for ch in [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]:
                xx,yy=ch[0],ch[1]
                if xx>=0 and xx<r and yy>=0 and yy<c:
                    if board[xx][yy] in t:
                        tt = board[x][y]
                        board[x][y]='*'
                        temp = board[xx][yy]
                        res = res+temp
                        dfs(xx,yy,t[temp],res)  
                        board[xx][yy]=temp
                        res = res[:-1]
                        board[x][y]=tt
            
                  
        
        ans = []
        
        for i in range(r):
            for j in range(c):
                if board[i][j] in trie:
                    dfs(i,j,trie[board[i][j]],board[i][j])
                
                    
        return ans
```

```python
class Solution:
    def grayCode(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        
        if n==0: return [0]
        
        res = []
        res.append('0')
        res.append('1')
        
        temp = res[:]
        ans = []
        
        for i in range(1,n):
            res = temp[:]
            temp=[]
            for j in range(len(res)):
                t = '0'+res[j]
                temp.append(t)
                
            for j in range(len(res)-1,-1,-1):
                t = '1'+res[j]
                temp.append(t)
        
        for t in temp:
            ans.append((int(t,2)))
        
        return ans
```

```python
class Solution:
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        
        res = []
        
        def backtrack(s, temp):
            if not s:
                res.append(temp[:])
                return
            
            for i in range(1,len(s)+1):
                if s[:i]==s[:i][::-1]:
                    backtrack(s[i:],temp+[s[:i]])
                    
        backtrack(s,[])
            
        return res
```

```python
class Solution:
    def floodFill(self, image, sr, sc, newColor):
        """
        :type image: List[List[int]]
        :type sr: int
        :type sc: int
        :type newColor: int
        :rtype: List[List[int]]
        """
        
        def dfs(i,j,r,c,seed,newColor):
            image[i][j]=newColor
            for xy in [(i+1,j),(i-1,j),(i,j+1),(i,j-1)]:
                x = xy[0]
                y = xy[1]
                
                if x>=0 and x<r and y>=0 and y<c and image[x][y]==seed and visited[x][y]==False:
                    visited[x][y]=True
                    dfs(x,y,r,c,seed,newColor)
                    visited[x][y]=False
                    
                    
        r = len(image)
        c = len(image[0])
        seed=image[sr][sc]
        visited = []
        for i in range(r):
            t = [False]*c
            visited.append(t)
        dfs(sr,sc,r,c,seed,newColor)
        
        return image
                    
```

```python
class Solution:
    def networkDelayTime(self, times, N, K):
        """
        :type times: List[List[int]]
        :type N: int
        :type K: int
        :rtype: int
        """
        
        dist = {}
        
        for i in range(1,N+1):
            for j in range(1,N+1):
                dist[(i,j)]=9999
        
        for time in times:
            i,j,t=time[0],time[1],time[2]
            dist[(i,j)]=t
        
        for k in range(1,N+1): 
            for i in range(1,N+1): 
                for j in range(1,N+1): 
                    dist[(i,j)] = min(dist[(i,j)],dist[(i,k)]+ dist[(k,j)])
        
        s = -1
        
        for i in range(1,N+1):
            if K!=i and dist[(K,i)]==9999: return -1
            if K!=i and dist[(K,i)]<9999:
                s=max(s,dist[(K,i)])
                    
        return s
        
        
```
