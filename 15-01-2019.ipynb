```python
class Solution(object):
    def largestPerimeter(self, A):
        """
        :type A: List[int]
        :rtype: int
        """
        
        A = sorted(A, reverse=True)
        
        l = len(A)
        
        for i in range(l-2):
            if A[i]<A[i+1]+A[i+2]:
                return A[i]+A[i+1]+A[i+2]
                    
        return 0
```

```python
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        n+=1
        m+=1
        val = [[0]*n]*m
        
        for i in range(1,m):
            for j in range(1,n):
                if i==1 or j==1:
                    val[i][j]=1
                else:
                    val[i][j]=val[i-1][j]+val[i][j-1]
            
        return val[-1][-1]
```

```python
class Solution(object):
    def uniquePathsWithObstacles(self, obstacleGrid):
        """
        :type obstacleGrid: List[List[int]]
        :rtype: int
        """
        
        if obstacleGrid[0][0]==1: 
            return 0
        
        obstacleGrid[0][0]=1
        
        n = len(obstacleGrid[0])
        m = len(obstacleGrid)
        
        for i in range(1,m):
            if obstacleGrid[i][0]==1 :
                obstacleGrid[i][0]=0
            else:
                obstacleGrid[i][0]=obstacleGrid[i-1][0]
                
        for j in range(1,n):
            if obstacleGrid[0][j]==1:
                obstacleGrid[0][j]=0
            else:
                obstacleGrid[0][j]=obstacleGrid[0][j-1]
                
        for i in range(1,m):
            for j in range(1,n):
                if obstacleGrid[i][j]==1:
                    obstacleGrid[i][j]=0
                else:
                    obstacleGrid[i][j]=obstacleGrid[i-1][j]+obstacleGrid[i][j-1]
             
        return obstacleGrid[-1][-1]
```

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        
        #nums = sorted(nums)
        cnt = len(nums)
        nums = sorted(nums)
        res = set([])
        
        for i, n in enumerate(nums):
            l = i+1
            r = cnt-1
            s = set()
            #print(l,r)
            while(l<r):
                s = n + nums[l]+nums[r]
                if s < 0:
                    l+=1
                elif s>0:
                    r-=1
                else:
                    res.add((n,nums[l],nums[r]))
                    l+=1
                    
        return list(res)               
```

```python
class Solution(object):
    def threeSumClosest(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        
        nums = sorted(nums)
        mn = float('Inf') 
        ans = target
        cnt =len(nums)
        
        for i,n in enumerate(nums):
            l = i+1
            r = cnt-1
            
            while(l<r):
                a = n + nums[l] + nums[r]
                #print(abs(target-a), mn, a, target)
                if a < target:
                    l+=1
                elif a > target:
                    r-=1
                else:
                    mn = a
                    ans = a
                    return ans
                #print(abs(target-a), mn)
                if abs(target-a)<mn:
                    mn = abs(target-a)
                    ans = a
                    
        return ans
                
        
```

```python
class Solution(object):
    def maxAreaOfIsland(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        
        self.mx = 0
        self.path = 0
        
        def dfs(grid,i,j,r,c):
            #print(i,j,self.path)
            self.path+=1
            if self.path > self.mx:
                self.mx = self.path
            for x,y in [(1,0),(-1,0),(0,1),(0,-1)]:
                #print(x,y)
                if x+i>=0 and y+j>=0 and x+i<r and y+j<c:
                    if grid[i+x][j+y]==1:
                        grid[i+x][j+y]=0
                        dfs(grid,i+x,j+y,r,c)
                        #grid[i][j]=1
                        
        r = len(grid)
        c = len(grid[0])
        #print(r,c)
        for i in range(r):
            for j in range(c):
                if grid[i][j]==1:
                    self.path = 0
                    grid[i][j]=0
                    dfs(grid,i,j,r,c)
                    #print(i,j,self.mx)
                    grid[i][j]=1
        return self.mx
```

```python
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """                    
        
        if grid==[]: return 0
        
        def dfs(grid, i, j, r, c): 
            #print(i,j,grid)
            grid[i][j]='#'
            for x,y, in [(i+1,j), (i-1,j), (i,j+1), (i,j-1)]:
                #print(x,y, grid)
                if x>=0 and y >=0 and x<r and y<c:
                    if grid[x][y]=="1":
                        dfs(grid, x, y, r, c)
            return
        
        r = len(grid)
        c = len(grid[0])
        
        cnt = 0
        for i in range(r):
            for j in range(c):
                if grid[i][j]=="1":
                    cnt+=1
                    dfs(grid, i,j,r,c)
                    
        return cnt
        
        
```

```python
class Solution:
    def subsetsWithDup(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        
        self.res = []
        
        nums = sorted(nums)
        def subset(nums, ind, temp):
            if temp not in self.res:
                self.res.append(temp)
                
            for i in range(ind,len(nums)):
                subset(nums,i+1,temp+[nums[i]])
                
        subset(nums,0,[])
        return self.res
            
```

```python
class Solution:
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        #here we need a global wise list, each time we just append to the result
        rslt=[]
        
        def dfs(temp, elements):
            #gather rslt
            print(temp)
            if len(elements)==0:
                rslt.append(temp[:]) #still remember to use temp[:]
            for e in elements:
                temp.append(e)
                #backtrack
                next_elements=elements[:]
                next_elements.remove(e)
                dfs(temp, next_elements)
                temp.pop()
                
                
        dfs([],nums) #first is the current result
        return rslt
```

```python
class Solution:
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        
        res = []
        
        def backtarck(temp, s, e, target):
            if target==0:
                res.append(temp[:])
            if target>0:
                for i in range(s,e):
                    temp.append(candidates[i])
                    backtarck(temp, i,e, target-candidates[i])
                    temp.pop()
                    
        
        candidates.sort(reverse=False)
        backtarck([],0, len(candidates), target)
            
        return res
```

```python
class Solution:
    def combinationSum2(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        
        res = []
        
        def backtrack(temp,s,e,target):
            if target==0:
                if temp[:] not in res:
                    res.append(temp[:])
            elif target > 0:
                for i in range(s,e):
                    temp.append(candidates[i])
                    backtrack(temp,i+1,e,target-candidates[i])
                    temp.pop()
                    
        candidates.sort(reverse=False)
        backtrack([],0,len(candidates), target)
        
        return res
```

```python
class Solution:
    def combinationSum3(self, k, n):
        """
        :type k: int
        :type n: int
        :rtype: List[List[int]]
        """
        
        res = []
        
        def backtrack(temp,s,e,cnt):
            if cnt==k and sum(temp)==n:
                res.append(temp[:])
            elif sum(temp) < n:
                for i in range(s,e):
                    if cnt>k: break
                    temp.append(candidates[i])
                    backtrack(temp,i+1,e,cnt+1)
                    temp.pop()
        
        candidates = [i for i in range(1,10)]
        backtrack([],0,len(candidates),0)
        
        return res
                
```
